**Gráfico de dispersión (scatter plot):**

Usando matplotlib y todas las especies tienen el mismo color

```python
plt.scatter(df['petal_length'], df['petal_width'])
# Personalizar
plt.title("Scatter Plot de Petal Length vs. Petal Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
```

Usando Seaborn y separado por color

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)
plt.title('Scatter Plot de petal Length y petal Width por Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

**d) Gráfico de cajas (box plot):**

Objetivo: Analizar la distribución de las medidas del sépalo (sepal_length, sepal_width).
Instrucciones: Crea un gráfico de cajas para mostrar la distribución de las variables sepal_length y sepal_width por especie.

```python
# Boxplot de Sepal Length por especie
sns.boxplot(x="species", y="sepal_length", data=df, hue='species', palette="Set2")
plt.title("Distribución de variables Sepal Length por especie")
plt.show()

# Boxplot de Sepal Width por especie
sns.boxplot(x="species", y="sepal_width", data=df, hue='species', palette='Set3')
plt.title("Distribución de variables Sepal Width por especie")
plt.show()
```
